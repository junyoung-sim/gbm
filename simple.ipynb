```python
import sys
import numpy as np
import pandas as pd
import seaborn as sns
import scipy as sp
import matplotlib.pyplot as plt
from download import download

np.set_printoptions(suppress=True)
```

```python
ticker = input("Ticker: ")
apikey = open("apikey", "r").readline()
download(ticker, apikey)
```

```python
dat = pd.read_csv("./data/{}.csv" .format(ticker))
dat = dat.loc[::-1].reset_index().drop(columns=["index"])
print(dat)
```

```python
returns = []
for t in range(1, dat.shape[0]):
    returns.append((dat["adjClose"][t] - dat["adjClose"][t-1]) * 100 / dat["adjClose"][t-1])
returns = np.array(returns)
returns = np.sort(returns)

print("{} P&L Statistics (%)" .format(ticker))
print(pd.DataFrame(returns).describe())
```

```python
ax = sns.histplot(data=returns, binwidth=np.std(returns), stat="probability")
ax.set(xlabel="{} P&L (%)" .format(ticker), ylabel="Frequency")
```

```python
x = np.linspace(np.min(returns), np.max(returns), 1000)
p = sp.stats.norm.pdf(x, returns.mean(), np.std(returns))
plt.plot(x, p)
plt.xlabel("{} P&L (%)" .format(ticker))
plt.ylabel("Gaussian Density")
```

```python
epoch = 10000
sample_size = 20
sample_path = []
sample_path_prob = []

for t in range(epoch):
    path = np.zeros(sample_size)
    path[0] = dat["adjClose"].iloc[-1]
    prob = np.ones(1)
    for i in range(1, sample_size):
        r = np.random.normal(returns.mean(), returns.std())
        path[i] = path[i-1] * (1.00 + r / 100)
        prob *= sp.stats.norm.pdf(r, returns.mean(), returns.std())
    sample_path.append(path)
    sample_path_prob.append(prob[0])

sample_path = np.array(sample_path)
sample_path_prob = np.array(sample_path_prob)
sample_path_prob = np.exp(sample_path_prob) / sum(np.exp(sample_path_prob))
```

```python
plt.figure()
for t in range(epoch):
    plt.plot(sample_path[t])
plt.show()
```

```python
for t in range(epoch):
    sample_path[t] *= sample_path_prob[t]
    
expected_value = sum(sample_path)
print(expected_value)
```
